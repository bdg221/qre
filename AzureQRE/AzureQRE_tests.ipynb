```python
from azure.quantum.qiskit import AzureQuantumProvider
from qiskit import QuantumCircuit, transpile
from qiskit.tools.monitor import job_monitor
```

```python
def create_CNOT():
    
    qc = QuantumCircuit(2)

    # Apply a CNOT gate with control on qubit 0 and target on qubit 1
    qc.cx(0, 1)
    qc.measure_all()


    basis_gates = backend.configuration().basis_gates
    circ = transpile(qc, basis_gates=basis_gates, optimization_level=0)

    print(f"[INFO]   qubit count: {circ.num_qubits}")
    print("[INFO]   gate counts")
    for gate, count in circ.count_ops().items():
        print(f"[INFO]   - {gate}: {count}")



    return circ
```

```python
def main(file_name, backend):
    qiskit_circuit = QuantumCircuit.from_qasm_file(file_name)
    job = backend.run(qiskit_circuit)
    job_monitor(job)
    result = job.result() 
    return result
```
